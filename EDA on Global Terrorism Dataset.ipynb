# Visualisation

```python
plt.figure(figsize=(12,8))
sns.countplot(df['iyear'])

plt.tight_layout()
plt.xticks(rotation=90)
plt.tick_params(axis='x',
    which='major',
    labelsize=8
    )
plt.tight_layout()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(
    x=df['attacktype1_txt'].value_counts().keys(),
    y=df['attacktype1_txt'].value_counts().values,
    )

plt.tight_layout()
plt.xticks(rotation=90)
plt.tick_params(axis='x',
                which='major',
                labelsize=8
                )
plt.tight_layout()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=df['region_txt'].value_counts().head(20).keys(),
            y=df['region_txt'].value_counts().head(20).values
           )

plt.tight_layout()
plt.xticks(rotation=90)
plt.tick_params(axis='x',
                which='major',
                labelsize=8
                )
plt.tight_layout()
```

```python
#Overall incidents 
plt.figure(figsize=(10,6))
sns.barplot(x=df['country_txt'].value_counts().head(10).keys(),
            y=df['country_txt'].value_counts().head(10).values,palette='magma'
           )
plt.tight_layout()
plt.xticks(rotation=90)
plt.tick_params(axis='x',
                which='major',
                labelsize=8
                )
plt.tight_layout()
```

# Conclusion:

The 80s and the 90s have been turbulent with on average 3000 incidents of terrorism each year. However, there has been a rapid increase in incidents post 2004.

Most of the incidents, over 80,000 have been **Bombings or Explosions**, followed by **Armed assault** with hijacking being the least common.

In terms of regions, **MEA (Middle East and North Africa)** has seen the most amount of violence, due to the presence of militant groups such as the Taliban, which are also responsible of carrying out the highest number of identified attacks followed by ISIL in the region.

The highest no. of incidents in any year has been in **2014 with 16903 cases**, with **Iraq being the country with the highest no. of incidents**, followed by **Pakistan and Afghanistan**. There have been **1249 incidents by ISIL**, which is due to the war in Iraq.

##### So to conclude, the MEA region, or specifically Iraq, is a hotspot for terrorist activites with the least being in Australia and Oceania region.
